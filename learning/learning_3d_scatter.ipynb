```python
import numpy as np
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d.axes3d as p3
import matplotlib.animation as animation
from IPython.display import display,HTML
from ipywidgets import IntProgress
import pandas as pd
```

```python
frame_num = 5 # 20 for fast generation, and 200 or more for smooth animation.
movie_length = 5000 # ms
```

```python
# animation's core function: generating every frame from 0 to frame_num-1
def update(frame):
    dot_size_max = 1000
    dot_size_min = 10
    dot_size = relative_helper(dot_size_min, dot_size_max, frame, (frame_num-1))
    g_dots.set_sizes([dot_size_min, dot_size, dot_size, dot_size_max])
    
    gPg.value = frame
    return g_dots,

def relative_helper(min=0, max=1., current=0, total=100):
    return current * (max-min) / total + min
```

```python
# prepare data
X = [1,2,3,4]
Y = [2,2,3,3]
Z = [0,0,1,2]
```

```python
# setting basic elements in the figure
fig = plt.figure(figsize=[8, 4.5])
g_ax = p3.Axes3D(fig)

g_dots = g_ax.scatter(X, Y, Z)
```

```python
gPg = IntProgress(min=0, max=frame_num-1, value=0)
display(gPg)
movie = animation.FuncAnimation(fig, update, frames=frame_num, interval=movie_length/frame_num, blit=True)
HTML(movie.to_html5_video())
```
