```python
import base
import fund_scanner.animation_framework.framework as fw

class MyDot(fw.FWDot):
    def __init__(self, offset=0, **kwargs):
        self.offset = offset
        return super().__init__(**kwargs)
    
    def step(self):
        self.size += 10
        return super().step()

class MyFancyDot(fw.FWGroup):
    def __init__(self, offset, **kwargs):
        self.offset = offset
        return super().__init__(**kwargs)

    def step(self):
        self.pos_x += 0.05
        if self.pos_x>self.movie.boundary('x_max'):
            self.pos_x=self.movie.boundary('x_max')
        self.pos_y += 0.05 + self.offset
        if self.pos_y>self.movie.boundary('y_max'):
            self.pos_y=self.movie.boundary('y_max')
        self.pos_z += (self.movie.boundary('z_max') - self.pos_z) / 20
        return super().step()
        

ani = fw.FWMovie(frame_num=500, figsize=[13,6], movie_length=20000)
dot = fw.FWDot(size=60, color='r', alpha=0.3)
dot1 = fw.FWDot(size=20)
t = fw.FWText(u'追')
g = MyFancyDot(0.2)
g.add([dot, dot1, t])



dot2 = MyDot(alpha=0.1)
t2 = fw.FWText(u'变大')
g2 = MyFancyDot(10)
g2.add([dot2, t2])
ani.add_objects([g, g2])

ani.display_movie(rotate=True)
```
