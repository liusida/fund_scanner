## 数据库设计

### 表. funds. 基金

funds_id 数据库自动编号

funds_code 基金代码 e.g.000199、000817、etc.

funds_name_full 基金全名

funds_name_short 基金简称

funds_type 基金类型 e.g. 定开债券、混合型、etc.

funds_start_date 成立日期

update_time 最新情况最近更新时间

historical_price_time 历史价格最近更新时间


### 表. funds_update. 基金最新情况

funds_update_id 数据库自动编号

funds_id 对应基金数据库编号

funds_price 当前基金单位净值

funds_price_adjust 当前基金累计净值

funds_amount 当前基金规模（单位亿）

funds_recent_1_month 基金最近1个月涨百分比

funds_recent_3_month 基金最近3个月涨百分比

funds_recent_6_month 基金最近6个月涨百分比

funds_recent_1_year 基金最近1年涨百分比

funds_recent_3_year 基金最近3年涨百分比

funds_return_total 基金成立来涨百分比

funds_return_adjust_per_year 基金成立来涨百分比（根据上证指数调整，并计算为年化）

funds_std_1_year 基金最近1年标准差

funds_std_2_year 基金最近2年标准差

funds_std_3_year 基金最近3年标准差

funds_sharp_1_year 基金最近1年夏普比率

funds_sharp_2_year 基金最近2年夏普比率

funds_sharp_3_year 基金最近3年夏普比率

funds_structure_personal 基金持有结构 个人

funds_structure_institutional 基金持有结构 机构

funds_structure_employee 基金持有结构 内部员工

funds_holding_concentration 基金持仓集中度

funds_holding_stock_percentage 基金持有股票占比

funds_holding_bond_percentage 基金持有债券占比

funds_holding_cash_percentage 基金持有现金占比


### 表. funds_holding. 基金持仓最新情况

funds_holding_id 数据库自动编号

funds_id 对应基金数据库编号

funds_holding_type 持仓是股票还是基金 0股票 1基金

funds_holding_code 持仓对象代码（股票代码或债券代码）

funds_holding_name 持仓对象名称

funds_holding_percentage 持仓对象占净值比重

funds_holding_amount 持仓对象市值


### 表. funds_historical_price. 基金历史价格

funds_historical_price_id 数据库自动编号

funds_id 对应基金编号

funds_price_date 日期

funds_price 基金单位净值

funds_price_adjust 基金累计净值（调整后净值）

funds_raising_percentage 基金当日增长

### 表. benchmarks. 对照标杆

benchmarks_id 数据库自动编号

benchmarks_name 标杆名称

historical_price_time 标杆历史价格最后更新时间

### 表. benchmarks_historical_price. 对照标杆的历史价格

benchmarks_historical_price_id 数据库自动编号

benchmarks_id 对应基金编号

benchmarks_price_date 日期

benchmarks_price 当日值

benchmarks_price_adjust 当日值（某种调整后值）



### 表. managers. 基金经理

managers_id 数据库自动编号

managers_name 姓名

managers_code 天天基金网编号

managers_photo 照片

managers_gendar_guess 猜测性别


### 表. companies. 基金公司

companies_id 数据库自动编号

companies_name_full 公司全称

companies_name_short 公司简称

companies_code 天天基金网编号

companies_address 公司地址

companies_start_date 公司成立时间

companies_size 公司规模


### 表. managers_2_funds. 基金经理对应基金

managers_2_funds_id 数据库自动编号

managers_id 基金经理数据库编号

funds_id 基金数据库编号

### 表. companies_2_funds. 基金公司对应基金

companies_2_funds_id 数据库自动编号

companies_id 基金公司数据库编号

funds_id 基金数据库编号

### 表. companies_2_mangers. 基金公司对应基金经理

companies_2_managers_id 数据库自动编号

companies_id 基金公司数据库编号

managers_id 基金经理编号

### 表. managers_2_funds_history. 基金经理对应基金的历史

managers_2_funds_history_id 数据库自动编号

managers_id 基金经理数据库编号

funds_id 基金数据库编号

history_start_date 时期开始日期（含）

history_end_date 时期结束日期（含）
